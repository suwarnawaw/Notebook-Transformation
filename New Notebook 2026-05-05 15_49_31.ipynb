```python
#Schema Definition
from pyspark.sql.types import StructType,StructField, StringType, IntegerType,DoubleType, BooleanType
admision_schema =  StructType([ \
    StructField("AdmissionId", StringType(), nullable = False),    
    StructField("PatientId", StringType(), nullable = False),
    StructField("HospitalId", StringType(), nullable = False),
    StructField("HospitalName", StringType(), nullable = False),
    StructField("DoctorId", StringType(), nullable = False),
    StructField("Department", StringType(), nullable = False),
    StructField("Ward", StringType(), nullable = False),
    StructField("BedNumber", StringType(), nullable = False),
    StructField("AdmissionType", StringType(), nullable = False),
    StructField("Status", StringType(), nullable = False)
  ])
#Read the data
df = spark.read.csv(Base_Path,header=True,schema=admision_schema)
df.createOrReplaceTempView("Admission")
display(df)
df.show()
df.printSchema()
df.count()
df.describe().show()
df.groupBy("AdmissionId").count().show()
df.show(3,truncate=False)
```

# Filtering Data

```python
# Filtering of data
df_critical = df_adm.filter(col('Status') == "Critical")
print('Critical admissions : ', df_critical.count())
df_critical = df_critical.select("AdmissionID","HospitalName","Department","Status","AdmissionType")
df_critical.show(5, truncate=False)
#Grouping of data
df_group = df_adm.groupBy("Department").count()
df_group.show(5, truncate=False)
```

```python
#Logical Operator : AND OR
#AND : If all conditions are true than true otherwise all are false
#OR : If any one condition is true than true otherwise all are false
#Filtering of data
df_critical = df_adm.filter((col('Status') == "Critical") & (col('AdmissionType') == "Emergency"))
print('Critical admissions : ', df_critical.count())
df_critical = df_critical.select("AdmissionID","HospitalName","Department","Status","AdmissionType")
df_critical.show()

#Grouping of data
df_group = df_adm.groupBy("Department").count()
df_group.show(5, truncate=False)

#Filtering of data
df_critical = df_adm.filter((col('Status') == "Critical") | (col('AdmissionType') == "Emergency"))
print('Critical admissions : ', df_critical.count())
df_critical = df_critical.select("AdmissionID","HospitalName","Department","Status","AdmissionType")
df_critical.show()

#Filtering of data
df_critical = df_adm.filter((col('Status') == "Critical") & (col('AdmissionType') == "Planned"))
print('Critical admissions : ', df_critical.count())
df_critical = df_critical.select("AdmissionID","HospitalName","Department","Status","AdmissionType")
df_critical.show()
```

```python
from pyspark.sql.functions import col, unix_timestamp, to_timestamp

# Read CSV
df_admission = spark.read.format("csv") \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .csv("/Volumes/workspace/default/sales_pyspark_session/admissions.csv")

# Show top 5 rows
df_admission.show(5, truncate=False)
df_admission.printSchema()
# Convert string column to timestamp
df_admission = df_admission.withColumn(
    "AdmissionTimeStamp",
    to_timestamp(col("LogTimestamp"), "yyyy-MM-dd'T'HH:mm:ss'Z'")
)

# Show converted data
df_admission.select("LogTimestamp","AdmissionTimeStamp").show  (5, truncate=False)

# Check schema
df_admission.printSchema()
```

```python
# Extracting date from Timestamp
from pyspark.sql.functions import col, unix_timestamp, to_timestamp, date_format
df_adm=spark.read.format("csv").option("header","true").option("inferSchema","true").csv("/Volumes/workspace/default/sales_pyspark_session/admissions.csv")
df_adm=df_adm.withColumn("AdmissionDate",date_format(col("LogTimestamp"),"yyyy-MM-dd"))
df_adm=df_adm.withColumn(
    "AdmissionTime",
    to_timestamp(col("LogTimestamp"), "'T'HH:mm:ss'Z'"))
df_adm = df_adm.select("AdmissionDate","AdmissionTime").show()
```

```python
# Rounding the Treatment Cost
from pyspark.sql.functions import col, unix_timestamp, to_timestamp, date_format, round

df_adm=spark.read.format("csv").option("header","true").option("inferSchema","true").csv("/Volumes/workspace/default/sales_pyspark_session/admissions.csv")
df_adm=df_adm.withColumn("TreatmentCost_INR",round(col("TreatmentCost_INR"),0))
df_adm.select("PatientID","TreatmentCost_INR").show(5, truncate=False)
```

```python
#Add col
from pyspark.sql.functions import (col,when)
df_adm=spark.read.format("csv").option("header","true").option("inferSchema","true").csv("/Volumes/workspace/default/sales_pyspark_session/admissions.csv")
df_adm=df_adm.withColumn("AdmissionType",when(col("AdmissionType")=="Urgent","Urgent").otherwise("Non-Urgent"))
df_adm.select("AdmissionID","AdmissionType").show(5, truncate=False)
#Add col
df_adm = spark.read.format("csv").option("header","true").option("inferSchema","true").csv("/Volumes/workspace/default/sales_pyspark_session/admissions.csv")
df_adm=df_adm.withColumn("TransactionType",when(col("TreatmentCost_INR")>100000,'Premium').when(col("TreatmentCost_INR")>50000,'Moderate').when(col("TreatmentCost_INR")>1000,'Normal'). otherwise('Low'))
df_adm.select("TreatmentCost_INR","TransactionType").show(5, truncate=False)
```

```python
# group by
df_adm = spark.read.format("csv").option("header","true").option("inferSchema","true").csv("/Volumes/workspace/default/sales_pyspark_session/admissions.csv")
df_adm=df_adm.groupBy("HospitalName").count().orderBy('count', ascending=False)

df_adm.show(5, truncate=False)
```

```python
from pyspark.sql.functions import col, unix_timestamp, to_timestamp, date_format, round

df_adm=spark.read.format("csv").option("header","true").option("inferSchema","true").csv("/Volumes/workspace/default/sales_pyspark_session/admissions.csv")
df_summary=df_adm.groupBy('HospitalName).agg(count('AdmissionID').alias("Total Admission").countDistinct('PatientID').alias("Total Patients").count(when(col('Status')=='Critical',1).allias("CriticalCases"))
,count(when(col('AdmissionType')=='Emergency',1).allias("Emergencies"), round(avg)("TreatmentCost_INR").alias("Avg Treatment Cost"),round(sum)("TreatmentCost_INR").alias("Total Treatment Cost")).orderBy('Total Admission', ascending=False)
df_summary.show(5, truncate=False)
```
